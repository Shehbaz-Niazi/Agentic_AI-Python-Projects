# **Python Assignment Operators**

```python
table[1]
```

```python
#  Addition Assigment Oprator (+=)

a: int = 15
b: int = 5

a += b
print(a)
a += 10
print(f"After adding b and 10, the final value of a is: {a}")

```

```python
#  	Subtraction assignment (-=)

a: int = 15
b: int = 5

a -= b
print(a)
a -= 5
print(f"After subtrating b and 5, the final value of a is: {a}")
```

```python
#  	Multiplication Assignment (*=)

a: int = 15
b: int = 2

a *= b
print(a)
a *= 5
print(f"After Multiplication b and 5, the final value of a is: {a}")

```

```python
#  	Division Assignment (/=)

a: int = 15
b: int = 5

a /= b
print(a)
a /= 3
print(f"After Division b and 3, the final value of a is: {a}")
```

```python
#   Modulus Assignment Operator (%=)

a: int = 17
b: int = 3

a %= b
print(a)
a %= 1
print(f"After Modulus b and 1, the final value of a is: {a}")
```

```python
#   Floor Division Assignment Operator (//=)

a: int = 17
b: int = 3

a //= b
print(a)

a //= 2
print(f"After Floor Division b and 2, the final value of a is: {a}")
```

```python
#   Exponentiation Assignment Operator (**=)

a: int = 2
b: int = 4

a **= b
print(a)

a **= 2
print(f"After Exponentiation b and 2, the final value of a is: {a}")
```

# **Python Logical Operators**

```python
table[3]
```

```python
#   Logical AND Operator (AND)

age: int = 20
student: str = True

if age >= 20 and student:
  print("You are allowed")
else:
  print("You are not Allowed")
```

```python
#   Logical OR Operator (OR)

age: int = 20
student: str = True

if age >= 15 or student:
  print("You are allowed")
else:
  print("You are not Allowed")
```

```python
#   Logical NOT Operator (NOT)

is_raining = True

if not is_raining:
    print("☀️ You can go outside!")
else:
    print("☔ Stay indoors")
```

```python
#   Identity Operator (is)

a: list[str] = ['a','b','c']
print(id(a))

b = a
print(id(b))

print(a is b)
```

# **Python Membership Operators**

```python
table[5]
```

```python
#   Membership Operator (in)

names: list[str] = ["shehbaz", "zia", "abubakar"]
print("shehbaz" in names)
print("haris" in names)
```

```python
#    Negative Membership Operator

names: list[str] = ["shehbaz", "zia", "abubakar"]
print("shehbaz" not in names)
print("haris" not in names)
```
